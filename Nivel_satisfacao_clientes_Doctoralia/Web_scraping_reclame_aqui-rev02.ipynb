```python
# Identifica a quantidade de páginas
#qtd_pag = int(driver.find_element(By.XPATH,'//*[@id="__next"]/div[1]/div[1]/div[3]/main/section[2]/div[2]/div[2]/div[11]/ul/li[8]').text)
qtd_pag = 50
#driver.close()

#//*[@id="__next"]/div[1]/div[1]/div[3]/main/section[2]/div[2]/div[2]/div[11]/ul/li[8]
#<li aria-label="pagination item" class="sc-cwhxAJ eKvhtt">226</li>
```

```python
rev_dict = {
    'Titulo': [],
    'Periodo': [],
    'Link': []
}

for pagina in range(1, qtd_pag + 1):
    
    url = 'https://www.reclameaqui.com.br/empresa/doctoralia/lista-reclamacoes/?pagina='+str(pagina)

    #Abre o link no Chrome e espera 10 segundos até tudo carregar
    driver = webdriver.Chrome()
    driver.get(url)
    time.sleep(120)

    driver.refresh()
    time.sleep(7)

    
    # Coleta o html da pagina 
    resposta = BeautifulSoup(driver.page_source, 'html.parser')
    # Cria uma lista com o código html de cada comentário (Nome, nota, período, review, links, fotos etc)
    reviews = resposta.find_all('div', class_='sc-1pe7b5t-0 iQGzPh')
    
    for result in reviews:
        titulo = result.find('h4', class_='sc-1pe7b5t-1 jAlTVn').text # encontra o código referente ao nome do usuário e salva em uma variável
        periodo = result.find('span',class_='sc-1pe7b5t-5 bmtSzo').text # encontra o código referente ao período do usuário e salva em uma variável
        link = result.find('a')['href']

        rev_dict['Titulo'].append(titulo)
        rev_dict['Periodo'].append(periodo)
        rev_dict['Link'].append('https://www.reclameaqui.com.br'+link)

    clear_output(wait = True)
    driver.close()

    print('avanço:'+str(pagina)+'/'+str(qtd_pag))
```

```python
df_reviews_reclame_aqui = pd.DataFrame(rev_dict) 
```

```python
df_reviews_reclame_aqui.to_excel("df_reclame_aqui.xlsx",sheet_name='reviews', index=True)
```

```python
rev_dict_detalhes = {
    'ID': [],
    #'Periodo_completo': [],
    'Data': [],
    'Cidade': [],
    'Estado': [],
    #'Local': [],
    'Review': [],
    'Categoria': [],
    'Produto': [],
    'Problema': [],
    #'Status': [],
}

contador = 0
for link in df_reviews_reclame_aqui['Link']:
        
    #Abre o link no Chrome e espera 10 segundos até tudo carregar
    driver = webdriver.Chrome()
    driver.get(link)
    time.sleep(7)

    # Coleta o html da pagina 
    resposta = BeautifulSoup(driver.page_source, 'html.parser')
    # Cria uma lista com o código html de cada comentário (Nome, nota, período, review, links, fotos etc)
    reviews = resposta.find_all('div', class_='lzlu7c-19 bewOsl')

    for result in reviews:
        id_ = result.find('span', class_='lzlu7c-12 ccjCCw').text.split()[1] # encontra o código referente ao nome do usuário e salva em uma variável
        #periodo_completo = result.find("span", {"data-testid": "complaint-creation-date"}).text # encontra o código referente ao período do usuário e salva em uma variável
        data = result.find("span", {"data-testid": "complaint-creation-date"}).text.split()[0] # encontra o código referente ao período do usuário e salva em uma variável
        cidade, estado = result.find("span", {"data-testid": "complaint-location"}).text.split(' - ')
        #local = result.find("span", {"data-testid": "complaint-location"}).text
        try:
            categoria = result.find('div', class_='sc-1s8uljb-0 FdBbS').text.split(', ')[0]
        except:
            categoria = np.nan
        try:
            produto = result.find('div', class_='sc-1s8uljb-0 FdBbS').text.split(', ')[1]
        except: 
            produto = categoria
        try:
            problema = result.find('div', class_='sc-1s8uljb-0 FdBbS').text.split(', ')[2]
        except:
            problema = categoria
        review_completo = result.find('p', class_='lzlu7c-17 cNqaUv').text
        #status = result.find('span', class_='sc-1a60wwz-1 kwzooO').text

        rev_dict_detalhes['ID'].append(id_)
        #rev_dict_detalhes['Periodo_completo'].append(periodo_completo)
        rev_dict_detalhes['Data'].append(data)
        #rev_dict_detalhes['Local'].append(local)
        rev_dict_detalhes['Cidade'].append(cidade)
        rev_dict_detalhes['Estado'].append(estado)
        rev_dict_detalhes['Review'].append(review_completo)
        rev_dict_detalhes['Categoria'].append(categoria)
        rev_dict_detalhes['Produto'].append(produto)
        rev_dict_detalhes['Problema'].append(problema)     
        #rev_dict_detalhes['Status'].append(status)

        driver.close()

    clear_output(wait = True)
    
    contador+=1
    print('avanço:'+str(contador)+'/'+str(len(df_reviews_reclame_aqui)))
```

```python
df_reviews_reclame_aqui_detalhes = pd.DataFrame(rev_dict_detalhes)
```

```python
df_completa_reclame_aqui = df_reviews_reclame_aqui.join(df_reviews_reclame_aqui_detalhes)
```
