```python
def trocar_textos(string_texto):
    # Retirando outros caracteres especiais
    texto = string_texto
    texto = texto.replace(',','')
    texto = texto.replace(';','')
    texto = texto.replace('!','')
    texto = texto.replace(':','')
    texto = texto.replace('.','')
    texto = texto.replace('"','')
    texto = texto.replace('(','')
    texto = texto.replace(')','')
    texto = texto.replace('[','')
    texto = texto.replace(']','')

    # Normalizando palavras gerais
    texto = texto.replace('medicos','medico')
    texto = texto.replace('medicas','medico')
    texto = texto.replace('medica','medico')
    texto = texto.replace('cancelar','cancelamento')
    texto = texto.replace('clientes','cliente')
    texto = texto.replace('pago','pagamento')
    texto = texto.replace('paguei','pagamento')
    texto = texto.replace('pagar','pagamento')
    texto = texto.replace('aplicativo','app')
    texto = texto.replace('plano','convenio')
    texto = texto.replace('avaliacoes','avaliacao')
    texto = texto.replace('consultas','consulta')
    texto = texto.replace('pacientes','paciente')
    texto = texto.replace('servicos','servico')
    texto = texto.replace('profissionais','profissional')
    texto = texto.replace('opinioes','opiniao')
    texto = texto.replace('opnioes','opiniao')
    texto = texto.replace('opniao','opiniao')
    return texto
```
